```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'Average_oil_price.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1', header=2)

# Rename the columns for better clarity
df.columns = ['Date', 'Crude_Petro', 'Crude_Brent', 'Crude_Dubai', 'Crude_WTI']

# Convert the 'Date' column to a datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%YM%m')

# Remove rows with missing or non-numeric data in 'Crude_Petro' and 'Crude_Brent' columns
df_cleaned = df.dropna(subset=['Crude_Petro', 'Crude_Brent', 'Crude_Dubai'])

# Plot with different colors for each price
plt.figure(figsize=(10, 6))

# Plot each series with a distinct color
plt.plot(df_cleaned['Date'], df_cleaned['Crude_Petro'], label='Crude Petro', linewidth=2, color='blue')
plt.plot(df_cleaned['Date'], df_cleaned['Crude_Brent'], label='Crude Brent', linewidth=2, color='green')
plt.plot(df_cleaned['Date'], df_cleaned['Crude_Dubai'], label='Crude Dubai', linewidth=2, color='red')

# Labeling the axes and title
plt.title('Crude Oil Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD per Barrel)')
plt.legend()

# Save the plot to a file
output_file_colored = 'crude_oil_price_over_time_colored.png'
plt.savefig(output_file_colored)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the coal price data
file_path_coal = 'Average_Coal_Price.xlsx'
df_coal = pd.read_excel(file_path_coal, sheet_name='Sheet2', header=2)

# Rename the columns for better clarity
df_coal.columns = ['Date', 'Coal_Australian', 'Coal_SouthAfrican', 'Unused']

# Remove the 'Unused' column
df_coal = df_coal.drop(columns=['Unused'])

# Convert the 'Date' column to a datetime format
df_coal['Date'] = pd.to_datetime(df_coal['Date'], format='%YM%m')

# Remove rows with non-numeric values in the 'Coal_Australian' and 'Coal_SouthAfrican' columns
df_coal['Coal_Australian'] = pd.to_numeric(df_coal['Coal_Australian'], errors='coerce')
df_coal['Coal_SouthAfrican'] = pd.to_numeric(df_coal['Coal_SouthAfrican'], errors='coerce')

# Drop rows with missing values
df_coal_cleaned = df_coal.dropna(subset=['Coal_Australian', 'Coal_SouthAfrican'])

# Calculate the Average Coal Price
df_coal_cleaned['Average_Coal_Price'] = df_coal_cleaned[['Coal_Australian', 'Coal_SouthAfrican']].mean(axis=1)

# Plot the coal prices over time
plt.figure(figsize=(10, 6))

# Plot each series with a distinct color
plt.plot(df_coal_cleaned['Date'], df_coal_cleaned['Coal_Australian'], label='Australian Coal', linewidth=2, color='blue')
plt.plot(df_coal_cleaned['Date'], df_coal_cleaned['Coal_SouthAfrican'], label='South African Coal', linewidth=2, color='green')
plt.plot(df_coal_cleaned['Date'], df_coal_cleaned['Average_Coal_Price'], label='Average Coal Price', linewidth=2, color='red')

# Labeling the axes and title
plt.title('Coal Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD per Metric Ton)')
plt.legend()



# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the natural gas price data
file_path_gas = 'Average_Gas_price.xlsx'
df_gas = pd.read_excel(file_path_gas, sheet_name='Sheet3', header=2)

# Rename the columns for better clarity
df_gas.columns = ['Date', 'Gas_US', 'Gas_Europe', 'Gas_Japan', 'Gas_Index', 'Average_Gas_Price']

# Remove the 'Gas_Index' column
df_gas = df_gas.drop(columns=['Gas_Index'])

# Convert the 'Date' column to a datetime format
df_gas['Date'] = pd.to_datetime(df_gas['Date'], format='%YM%m')

# Remove rows with non-numeric values in the relevant columns
df_gas['Gas_US'] = pd.to_numeric(df_gas['Gas_US'], errors='coerce')
df_gas['Gas_Europe'] = pd.to_numeric(df_gas['Gas_Europe'], errors='coerce')
df_gas['Gas_Japan'] = pd.to_numeric(df_gas['Gas_Japan'], errors='coerce')
df_gas['Average_Gas_Price'] = pd.to_numeric(df_gas['Average_Gas_Price'], errors='coerce')

# Drop rows with missing values
df_gas_cleaned = df_gas.dropna(subset=['Gas_US', 'Gas_Europe', 'Gas_Japan', 'Average_Gas_Price'])

# Plot the natural gas prices over time
plt.figure(figsize=(10, 6))

# Plot each series with a distinct color
plt.plot(df_gas_cleaned['Date'], df_gas_cleaned['Gas_US'], label='US Natural Gas', linewidth=2, color='blue')
plt.plot(df_gas_cleaned['Date'], df_gas_cleaned['Gas_Europe'], label='Europe Natural Gas', linewidth=2, color='green')
plt.plot(df_gas_cleaned['Date'], df_gas_cleaned['Gas_Japan'], label='Japan LNG', linewidth=2, color='red')
plt.plot(df_gas_cleaned['Date'], df_gas_cleaned['Average_Gas_Price'], label='Average Gas Price', linewidth=2, color='purple')

# Labeling the axes and title
plt.title('Natural Gas Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD per MMBtu)')
plt.legend()


# Show the plot
plt.show()
```
